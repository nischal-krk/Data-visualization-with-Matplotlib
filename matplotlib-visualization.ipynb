```python
import matplotlib.pyplot as plt                          #importing packages and subpackages
```

```python
mount = ['Eve','Kan','Lho','Mak','Cho','Dhau-I','Man']   #first dataset
height = [8848,8586,8516,8463,8201,8167,8156]            #second dataset
```

```python
plt.plot(mount,height)                                   #code to display line plot
```

```python
plt.scatter(mount,height)                                  #code to display scatter plot
plt.xlabel('Mountains')                                    #setting label on x-axis
plt.ylabel('Height in Metres')                             #setting label on y-axis
plt.title('7 Highest Mountains of Nepal.')                 #setting title of the plot
```

```python
plt.hist(height,bins=5)                                     #code to display histogram
```
